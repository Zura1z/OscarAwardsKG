```python
f = open('the_oscar_award2.csv')
reader = csv.reader(f)
OscarOntology = Namespace(onto.base_iri)
g = Graph()
g.bind("OscarOntology", OscarOntology)
g.bind("foaf", FOAF)
g.bind("xsd",XSD)
g.bind("iri",iri)
with onto:
    i=0
    for row in reader:
        i+=1
        if i==10:
                
                break
        rowNo,csv_year_film, csv_year_ceremomy, csv_ceremony,csv_category, csv_name, csv_name_film, csv_winner = row
        
        csv_name = csv_name.replace('"','')
        csv_name = csv_name.replace("'",'')
        csv_name = csv_name.replace('.','')
        csv_name = csv_name.replace(';','')
        csv_name = csv_name.replace('!','')
        csv_name = csv_name.replace(':','')
        csv_name = csv_name.replace(')','')
        csv_name = csv_name.replace('[','')
        csv_name = csv_name.replace(']','')
        csv_name = csv_name.replace('(','')
        csv_name = csv_name.replace('Ã','A')
        csv_name = csv_name.replace('ª','')
        csv_name = csv_name.replace(' and ',', ')
        if csv_name
        csv_name=csv_name.split(', ')
        
        
        #print(csv_name,len(csv_name))
        
        csv_category = csv_category.replace('"','')
        csv_category = csv_category.replace('.','')
        csv_category = csv_category.replace("'",'')
        csv_category = csv_category.replace(',','')
        csv_category = csv_category.replace(';','')
        csv_category = csv_category.replace('!','')
        csv_category = csv_category.replace(':','')
        csv_category = csv_category.replace(')','')
        csv_category = csv_category.replace('(','')
        csv_category = csv_category.replace('[','')
        csv_category = csv_category.replace(']','')
        csv_category = csv_category.replace('--','_')
        csv_category = csv_category.replace('-','_')
        csv_category = csv_category.replace('Ã','A')
        csv_category = csv_category.replace('ª','')
        
        
        csv_name_film = csv_name_film.replace('"','')
        csv_name_film = csv_name_film.replace('.','')
        csv_name_film = csv_name_film.replace("'",'')
        csv_name_film = csv_name_film.replace(',','')
        csv_name_film = csv_name_film.replace('!','')
        csv_name_film = csv_name_film.replace(';','')
        csv_name_film = csv_name_film.replace(':','')
        csv_name_film = csv_name_film.replace(')','')
        csv_name_film = csv_name_film.replace('(','')
        csv_name_film = csv_name_film.replace('[','')
        csv_name_film = csv_name_film.replace(']','')
        csv_name_film = csv_name_film.replace('--','_')
        csv_name_film = csv_name_film.replace('-','_')
        csv_name_film = csv_name_film.replace('Ã','A')
        csv_name_film = csv_name_film.replace('ª','')
        
        
     
        
        film_name = Literal(csv_name_film,datatype=XSD.string)
        ceremomy_year = Literal(csv_year_ceremomy,datatype=XSD.int)
        ceremomy_No = Literal(csv_ceremony,datatype=XSD.int)
        category = Literal(csv_category,datatype=XSD.string)
        nominee_name = Literal(csv_name,datatype=XSD.string)
        film_year = Literal(csv_year_film,datatype=XSD.int)
        winner = Literal(csv_winner,datatype=XSD.string)
        
        ceremomy_iri = URIRef((OscarOntology+ceremomy_year.replace(' ', '_'))) 
        category_iri = URIRef((OscarOntology+'cat/'+category.replace(' ', '_')))
        if len(csv_name) > 1:
            for name in csv_name:
                nominee_iri = URIRef((OscarOntology+'nominee/'+name.replace(' ', '_')))
                
        else:
            nominee_iri = URIRef((OscarOntology+'nominee/'+nominee_name.replace(' ', '_')))
        
        film_iri = URIRef((OscarOntology+'film/'+film_name.replace(' ', '_'))) 
        g.add((ceremomy_iri,RDF.type,iri.ceremomy))
        g.add((ceremomy_iri,iri.hasCeremonyYear,ceremomy_year))
        g.add((ceremomy_iri,iri.hasCeremonyNo,ceremomy_No))
        g.add((ceremomy_iri,iri.ceremonyConsistsOf,category_iri))
        
        g.add((category_iri,RDF.type,iri.category))
        
        g.add((nominee_iri,RDF.type,iri.nominee))
        g.add((nominee_iri,iri.nomineeBelongsTo,film_iri))
        g.add((nominee_iri,iri.hasBeenNominated,ceremomy_iri))
        g.add((nominee_iri,iri.hasCategories,category_iri))
        
        g.add((nominee_iri,iri.hasNomineeName,nominee_name))
        
        if csv_winner=="True":
            g.add((nominee_iri,iri.hasResult,iri.Winner))
        else:
            g.add((nominee_iri,iri.hasResult,iri.Losser))
        
        g.add((film_iri,RDF.type,iri.film))
        g.add((film_iri,iri.hasFilmName,film_name))
        g.add((film_iri,iri.hasfilmYear,film_year))
```

```python
g.serialize(destination='oscarv13.ttl', format='ttl')
```
